## 1.Use IF ELSE and ELIF to write a program in python for your Report Cards.

```python
maths=int(input("Enter maths marks"))
physics=int(input("Enter physics marks"))
chemistry=int(input("Enter chemistry marks"))
per=((maths+physics+chemistry)/300)*100
if (per>=90):
    print("A grade")
elif ((per>=80)and(per<90)):
    print("B grade")
elif ((per>=70)and(per<80)):
    print("C grade")
elif ((per>=60)and(per<70)):
    print("D grade")
elif ((per>=50)and(per<60)):
    print("E grade")
else:
    print("fail")
```

## 2.Use For Loop to Print Prime Numbers in between 1 to 1000


```python
for i in range(1,1001):
    if(i>1):
        for j in range(2,i):
            if(i%j==0):
                break
        else:
            print(i,end=" ")
```

## 3.Write a program for printing the tables from 1,10 using Nested For Loop

```python
for i in range(1,11):
    for j in range(1,11):
        print("{}x{}={}\n".format(i,j,i*j))
    print("\t")    
```
